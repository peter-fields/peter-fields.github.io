# Attention Diagnostics: (KL, Var) Phase Space for IOI Circuit

Computes two diagnostics for each attention head in GPT-2-small:
- **Normalized KL**: $\hat\rho_{\text{eff}} / \log n = (\log n - H(\hat\pi)) / \log n$
- **Normalized Var**: $\text{Var}_{\hat\pi}(\log \hat\pi) / (\log n)^2$

Tests against the IOI circuit from Wang et al. 2022.

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from transformer_lens import HookedTransformer

torch.set_grad_enabled(False)

model = HookedTransformer.from_pretrained("gpt2-small")
print(f"Loaded: {model.cfg.n_layers} layers, {model.cfg.n_heads} heads/layer")
```

## IOI Circuit Head Labels
From Wang et al. 2022 (arXiv:2211.00593)

```python
# IOI circuit head labels: (layer, head)
IOI_HEADS = {
    "Name Mover": [(9, 6), (9, 9), (10, 0)],
    "Backup Name Mover": [(9, 0), (9, 7), (10, 1), (10, 2), (10, 6), (10, 10), (11, 2), (11, 9)],
    "Negative Name Mover": [(10, 7), (11, 10)],
    "S-Inhibition": [(7, 3), (7, 9), (8, 6), (8, 10)],
    "Induction": [(5, 5), (6, 9)],
    "Duplicate Token": [(0, 1), (3, 0)],
    "Previous Token": [(2, 2), (4, 11)],
}

# Flat set for quick lookup
ALL_CIRCUIT_HEADS = set()
for heads in IOI_HEADS.values():
    ALL_CIRCUIT_HEADS.update(heads)

def head_role(layer, head):
    """Return circuit role string, or 'Non-circuit' if not in IOI circuit."""
    for role, heads in IOI_HEADS.items():
        if (layer, head) in heads:
            return role
    return "Non-circuit"

print(f"Total circuit heads: {len(ALL_CIRCUIT_HEADS)}")
for role, heads in IOI_HEADS.items():
    print(f"  {role}: {heads}")
```

## IOI Prompt Generation

```python
import random

# IOI templates (ABBA and BABA patterns) — one name repeats as subject
TEMPLATES_ABBA = [
    "When {A} and {B} went to the store, {B} gave a drink to",
    "When {A} and {B} went to the park, {B} handed a ball to",
    "When {A} and {B} arrived at the office, {B} passed a note to",
    "When {A} and {B} got to the restaurant, {B} offered a menu to",
    "When {A} and {B} walked into the room, {B} showed a book to",
    "After {A} and {B} met at the cafe, {B} sent a message to",
    "After {A} and {B} sat down for dinner, {B} gave a gift to",
]

TEMPLATES_BABA = [
    "When {B} and {A} went to the store, {B} gave a drink to",
    "When {B} and {A} went to the park, {B} handed a ball to",
    "When {B} and {A} arrived at the office, {B} passed a note to",
    "When {B} and {A} got to the restaurant, {B} offered a menu to",
    "When {B} and {A} walked into the room, {B} showed a book to",
    "After {B} and {A} met at the cafe, {B} sent a message to",
    "After {B} and {A} sat down for dinner, {B} gave a gift to",
]

# Non-IOI controls: same structure, same length, but the subject of the second
# clause is a THIRD name C (not A or B). No name repeats → IOI circuit shouldn't fire.
TEMPLATES_NON_IOI = [
    "When {A} and {B} went to the store, {C} gave a drink to",
    "When {A} and {B} went to the park, {C} handed a ball to",
    "When {A} and {B} arrived at the office, {C} passed a note to",
    "When {A} and {B} got to the restaurant, {C} offered a menu to",
    "When {A} and {B} walked into the room, {C} showed a book to",
    "After {A} and {B} met at the cafe, {C} sent a message to",
    "After {A} and {B} sat down for dinner, {C} gave a gift to",
]

NAMES = [
    "Mary", "John", "Alice", "Bob", "Sarah", "Tom",
    "Emma", "James", "Lisa", "David", "Kate", "Mark",
]


def generate_unique_prompts_ioi(n, seed=42):
    """Generate n unique IOI prompts (one name repeats)."""
    random.seed(seed)
    templates = TEMPLATES_ABBA + TEMPLATES_BABA
    seen = set()
    prompts = []
    while len(prompts) < n:
        template = random.choice(templates)
        a, b = random.sample(NAMES, 2)
        prompt = template.format(A=a, B=b)
        if prompt not in seen:
            seen.add(prompt)
            prompts.append(prompt)
    return prompts


def generate_unique_prompts_non_ioi(n, seed=43):
    """Generate n unique non-IOI prompts (third name C, no repetition)."""
    random.seed(seed)
    seen = set()
    prompts = []
    while len(prompts) < n:
        template = random.choice(TEMPLATES_NON_IOI)
        a, b, c = random.sample(NAMES, 3)
        prompt = template.format(A=a, B=b, C=c)
        if prompt not in seen:
            seen.add(prompt)
            prompts.append(prompt)
    return prompts


ioi_prompts = generate_unique_prompts_ioi(50)
non_ioi_prompts = generate_unique_prompts_non_ioi(50)

assert len(set(ioi_prompts)) == 50, "IOI prompts have duplicates!"
assert len(set(non_ioi_prompts)) == 50, "Non-IOI prompts have duplicates!"

print(f"IOI prompts: {len(ioi_prompts)} (all unique)")
for i in range(3):
    print(f"  {ioi_prompts[i]}")
print(f"\nNon-IOI prompts: {len(non_ioi_prompts)} (all unique)")
for i in range(3):
    print(f"  {non_ioi_prompts[i]}")

# Verify lengths match
ioi_lens = [model.to_tokens(p).shape[1] for p in ioi_prompts]
non_ioi_lens = [model.to_tokens(p).shape[1] for p in non_ioi_prompts]
print(f"\nToken lengths — IOI: {min(ioi_lens)}-{max(ioi_lens)} (mean {np.mean(ioi_lens):.1f}), "
      f"Non-IOI: {min(non_ioi_lens)}-{max(non_ioi_lens)} (mean {np.mean(non_ioi_lens):.1f})")
```

## Diagnostic Computation

```python
def compute_diagnostics(model, prompts):
    """
    For each prompt, compute normalized KL and normalized Var for all heads
    at the final token position.
    
    Returns:
        kl: (n_prompts, n_layers, n_heads)
        var: (n_prompts, n_layers, n_heads)
    """
    n_layers = model.cfg.n_layers
    n_heads = model.cfg.n_heads
    n_prompts = len(prompts)
    
    kl_all = np.zeros((n_prompts, n_layers, n_heads))
    var_all = np.zeros((n_prompts, n_layers, n_heads))
    
    for p_idx, prompt in enumerate(prompts):
        tokens = model.to_tokens(prompt)  # (1, seq_len)
        _, cache = model.run_with_cache(tokens)
        
        seq_len = tokens.shape[1]
        log_n = np.log(seq_len)
        
        for layer in range(n_layers):
            # attention pattern: (1, n_heads, seq_len, seq_len)
            attn = cache["pattern", layer]  # (1, n_heads, dest, src)
            # Take final token's attention over all source positions
            pi = attn[0, :, -1, :]  # (n_heads, seq_len)
            
            log_pi = torch.log(pi + 1e-12)
            
            # H(pi) = -sum pi * log pi
            entropy = -(pi * log_pi).sum(dim=-1)  # (n_heads,)
            
            # KL(pi || u) = log n - H(pi), normalized by log n
            kl_norm = (log_n - entropy.cpu().numpy()) / log_n
            
            # Var_pi(log pi), normalized by (log n)^2
            mean_log_pi = (pi * log_pi).sum(dim=-1, keepdim=True)  # (n_heads, 1)
            var_log_pi = (pi * (log_pi - mean_log_pi) ** 2).sum(dim=-1)  # (n_heads,)
            var_norm = var_log_pi.cpu().numpy() / (log_n ** 2)
            
            kl_all[p_idx, layer, :] = kl_norm
            var_all[p_idx, layer, :] = var_norm
    
    return kl_all, var_all


print("Computing diagnostics on IOI prompts...")
kl_ioi, var_ioi = compute_diagnostics(model, ioi_prompts)
print(f"  Shape: {kl_ioi.shape}")

print("Computing diagnostics on non-IOI prompts...")
kl_non_ioi, var_non_ioi = compute_diagnostics(model, non_ioi_prompts)
print(f"  Shape: {kl_non_ioi.shape}")

print("Done!")
```

## Exploration: Is χ redundant with KL? And does ⟨z⟩ do better?

Check (1) correlation between KL and χ across heads/prompts, and (2) whether excess expected score ⟨z⟩_π − ⟨z⟩_u provides an independent second axis.

```python
def compute_all_diagnostics(model, prompts):
    """
    Compute KL, chi (Var), and excess expected score for all heads.
    Returns arrays of shape (n_prompts, n_layers, n_heads).
    """
    n_layers = model.cfg.n_layers
    n_heads = model.cfg.n_heads
    n_prompts = len(prompts)

    kl_all = np.zeros((n_prompts, n_layers, n_heads))
    chi_all = np.zeros((n_prompts, n_layers, n_heads))
    ez_excess_all = np.zeros((n_prompts, n_layers, n_heads))

    for p_idx, prompt in enumerate(prompts):
        tokens = model.to_tokens(prompt)
        _, cache = model.run_with_cache(tokens)

        seq_len = tokens.shape[1]
        log_n = np.log(seq_len)

        for layer in range(n_layers):
            # Attention weights: (1, n_heads, seq, seq)
            pi = cache["pattern", layer][0, :, -1, :]  # (n_heads, seq_len)

            # Raw scores (pre-softmax): TransformerLens stores "attn_scores"
            # shape: (1, n_heads, seq, seq)
            z = cache["attn_scores", layer][0, :, -1, :]  # (n_heads, seq_len)

            log_pi = torch.log(pi + 1e-12)

            # KL = log n - H(pi), normalized
            entropy = -(pi * log_pi).sum(dim=-1)
            kl_norm = (log_n - entropy.cpu().numpy()) / log_n

            # chi = Var_pi(log pi), normalized
            mean_log_pi = (pi * log_pi).sum(dim=-1, keepdim=True)
            var_log_pi = (pi * (log_pi - mean_log_pi) ** 2).sum(dim=-1)
            chi_norm = var_log_pi.cpu().numpy() / (log_n ** 2)

            # Excess expected score: <z>_pi - <z>_u
            ez_pi = (pi * z).sum(dim=-1)          # (n_heads,)
            ez_u = z.mean(dim=-1)                  # (n_heads,)
            ez_excess = (ez_pi - ez_u).cpu().numpy()

            kl_all[p_idx, layer, :] = kl_norm
            chi_all[p_idx, layer, :] = chi_norm
            ez_excess_all[p_idx, layer, :] = ez_excess

    return kl_all, chi_all, ez_excess_all


print("Computing full diagnostics on IOI prompts...")
kl_ioi_full, chi_ioi, ez_ioi = compute_all_diagnostics(model, ioi_prompts)
print("Computing full diagnostics on non-IOI prompts...")
kl_nonioi_full, chi_nonioi, ez_nonioi = compute_all_diagnostics(model, non_ioi_prompts)
print("Done!")
```

## Where do non-activated circuit heads actually attend?

Verify directly: when circuit heads have high KL on non-IOI prompts, what tokens are they peaked on?

```python
# For a selection of circuit heads, look at where they attend on non-IOI prompts
# For each head & prompt: find the argmax token and the top-3 tokens with their weights

HEADS_TO_INSPECT = [
    ((9, 9),  "Name Mover"),
    ((9, 6),  "Name Mover"),
    ((7, 3),  "S-Inhibition"),
    ((5, 5),  "Induction"),
    ((3, 0),  "Duplicate Token"),
    ((4, 11), "Previous Token"),
]

print("=== Non-IOI prompts: where do circuit heads attend? ===\n")

for (l, h), role in HEADS_TO_INSPECT:
    # Collect argmax position stats across all non-IOI prompts
    argmax_positions = []  # relative position (0 = first token, -1 = last)
    argmax_tokens = []
    max_weights = []

    for prompt in non_ioi_prompts[:20]:  # first 20 for readability
        tokens = model.to_tokens(prompt)
        _, cache = model.run_with_cache(tokens)
        pi = cache["pattern", l][0, h, -1, :]  # (seq_len,)
        seq_len = tokens.shape[1]

        # Top-3 positions
        topk = torch.topk(pi, k=min(3, seq_len))
        top_positions = topk.indices.cpu().numpy()
        top_weights = topk.values.cpu().numpy()

        # Decode tokens at those positions
        token_strs = [model.to_string(tokens[0, pos:pos+1]) for pos in top_positions]

        argmax_positions.append(top_positions[0])
        argmax_tokens.append(token_strs[0])
        max_weights.append(top_weights[0])

    # Summary
    argmax_positions = np.array(argmax_positions)
    max_weights = np.array(max_weights)

    # Count how often argmax is position 0 (BOS)
    bos_frac = (argmax_positions == 0).mean()
    # Count how often argmax is the last or second-to-last position
    # (we look at attention FROM the last token, so prev = seq_len - 2)

    print(f"L{l}H{h} ({role}):")
    print(f"  Argmax is BOS (pos 0):  {bos_frac:.0%}")
    print(f"  Mean max weight:        {max_weights.mean():.3f} ± {max_weights.std():.3f}")
    print(f"  Most common argmax tokens: ", end="")

    from collections import Counter
    tok_counts = Counter(argmax_tokens)
    for tok, cnt in tok_counts.most_common(5):
        print(f"'{tok.strip()}'({cnt})", end="  ")
    print(f"\n  Argmax positions: {argmax_positions.tolist()}")
    print()
```

## Formal test: is |ΔKL| larger for circuit heads?

```python
from scipy import stats

abs_dkl_circuit = []
abs_dkl_noncircuit = []

for layer in range(12):
    for head in range(12):
        val = abs(delta_kl[layer, head])
        if (layer, head) in ALL_CIRCUIT_HEADS:
            abs_dkl_circuit.append(val)
        else:
            abs_dkl_noncircuit.append(val)

abs_dkl_circuit = np.array(abs_dkl_circuit)
abs_dkl_noncircuit = np.array(abs_dkl_noncircuit)

# Mann-Whitney U test (non-parametric, doesn't assume normality)
u_stat, p_val = stats.mannwhitneyu(abs_dkl_circuit, abs_dkl_noncircuit, alternative='greater')

print(f"=== |ΔKL| comparison: circuit vs non-circuit heads ===")
print(f"  Circuit heads (n={len(abs_dkl_circuit)}):     mean |ΔKL| = {abs_dkl_circuit.mean():.4f} ± {abs_dkl_circuit.std():.4f}")
print(f"  Non-circuit heads (n={len(abs_dkl_noncircuit)}): mean |ΔKL| = {abs_dkl_noncircuit.mean():.4f} ± {abs_dkl_noncircuit.std():.4f}")
print(f"  Mann-Whitney U (circuit > non-circuit): U = {u_stat:.0f}, p = {p_val:.4f}")
print()

# Also do the same for |Δχ|
abs_dchi_circuit = np.array([abs(delta_chi[l, h]) for l, h in ALL_CIRCUIT_HEADS])
abs_dchi_noncircuit = np.array([abs(delta_chi[l, h]) for l in range(12) for h in range(12)
                                 if (l, h) not in ALL_CIRCUIT_HEADS])

u_stat2, p_val2 = stats.mannwhitneyu(abs_dchi_circuit, abs_dchi_noncircuit, alternative='greater')

print(f"=== |Δχ| comparison: circuit vs non-circuit heads ===")
print(f"  Circuit heads:     mean |Δχ| = {abs_dchi_circuit.mean():.4f} ± {abs_dchi_circuit.std():.4f}")
print(f"  Non-circuit heads: mean |Δχ| = {abs_dchi_noncircuit.mean():.4f} ± {abs_dchi_noncircuit.std():.4f}")
print(f"  Mann-Whitney U (circuit > non-circuit): U = {u_stat2:.0f}, p = {p_val2:.4f}")

# Histogram
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

ax = axes[0]
ax.hist(abs_dkl_noncircuit, bins=20, alpha=0.5, color="#cccccc", label="Non-circuit", density=True)
ax.hist(abs_dkl_circuit, bins=12, alpha=0.7, color="#d62728", label="Circuit", density=True)
ax.set_xlabel("|ΔKL| (IOI − non-IOI)", fontsize=11)
ax.set_ylabel("Density", fontsize=11)
ax.set_title(f"|ΔKL| distribution (p = {p_val:.4f})", fontsize=12)
ax.legend()
ax.grid(True, alpha=0.2)

ax = axes[1]
ax.hist(abs_dchi_noncircuit, bins=20, alpha=0.5, color="#cccccc", label="Non-circuit", density=True)
ax.hist(abs_dchi_circuit, bins=12, alpha=0.7, color="#d62728", label="Circuit", density=True)
ax.set_xlabel("|Δχ| (IOI − non-IOI)", fontsize=11)
ax.set_ylabel("Density", fontsize=11)
ax.set_title(f"|Δχ| distribution (p = {p_val2:.4f})", fontsize=12)
ax.legend()
ax.grid(True, alpha=0.2)

plt.tight_layout()
plt.savefig("delta_distributions.png", dpi=150, bbox_inches="tight")
plt.show()
```

```python
# === Cross-head KL covariance under perturbation ===
# For all 50x50 = 2500 pairings of IOI prompt i with non-IOI prompt j,
# compute ΔKL(i,j,head) = KL_IOI(i,head) - KL_nonIOI(j,head).
# Then compute the 144x144 correlation of those ΔKL vectors.
#
# Also compute C_IOI, C_nonIOI, and C_IOI - C_nonIOI separately.

# Reshape: (50, 144)
kl_ioi_mat = kl_ioi.reshape(50, -1)
kl_non_mat = kl_non_ioi.reshape(50, -1)

# Correlation matrices on each condition
C_ioi = np.corrcoef(kl_ioi_mat.T)    # (144, 144)
C_non = np.corrcoef(kl_non_mat.T)    # (144, 144)
C_diff = C_ioi - C_non               # what changes

# All-pairings ΔKL covariance
delta_all = np.zeros((2500, 144))
idx = 0
for i in range(50):
    for j in range(50):
        delta_all[idx] = kl_ioi_mat[i] - kl_non_mat[j]
        idx += 1
C_delta = np.corrcoef(delta_all.T)   # (144, 144)

# Order by circuit role
role_order = ["Name Mover", "Backup Name Mover", "Negative Name Mover",
              "S-Inhibition", "Induction", "Duplicate Token", "Previous Token"]

ordered_indices = []
ordered_labels = []
ordered_roles = []
for role in role_order:
    for (l, h) in IOI_HEADS[role]:
        ordered_indices.append(l * 12 + h)
        ordered_labels.append(f"L{l}H{h}")
        ordered_roles.append(role)
n_circuit = len(ordered_indices)
for l in range(12):
    for h in range(12):
        if (l, h) not in ALL_CIRCUIT_HEADS:
            ordered_indices.append(l * 12 + h)
            ordered_labels.append(f"L{l}H{h}")
            ordered_roles.append("Non-circuit")
ordered_indices = np.array(ordered_indices)

# Reorder all matrices
C_ioi_ord = C_ioi[np.ix_(ordered_indices, ordered_indices)]
C_non_ord = C_non[np.ix_(ordered_indices, ordered_indices)]
C_diff_ord = C_diff[np.ix_(ordered_indices, ordered_indices)]
C_delta_ord = C_delta[np.ix_(ordered_indices, ordered_indices)]

# Role boundaries
role_boundaries = []
cumulative = 0
for role in role_order:
    cumulative += len(IOI_HEADS[role])
    role_boundaries.append(cumulative)

def plot_corr_matrix(ax, mat, title, vmin=-0.5, vmax=0.5, cmap='RdBu_r'):
    im = ax.imshow(mat, cmap=cmap, vmin=vmin, vmax=vmax, aspect='equal')
    ax.axhline(n_circuit - 0.5, color='black', lw=1.5)
    ax.axvline(n_circuit - 0.5, color='black', lw=1.5)
    for b in role_boundaries[:-1]:
        ax.axhline(b - 0.5, color='gray', lw=0.5, ls='--')
        ax.axvline(b - 0.5, color='gray', lw=0.5, ls='--')
    ax.set_title(title, fontsize=9)
    return im

# === Full 144x144 ===
fig, axes = plt.subplots(2, 2, figsize=(16, 14))

im1 = plot_corr_matrix(axes[0,0], C_ioi_ord, "C_IOI: KL corr on IOI prompts")
plt.colorbar(im1, ax=axes[0,0], shrink=0.7)

im2 = plot_corr_matrix(axes[0,1], C_non_ord, "C_nonIOI: KL corr on non-IOI prompts")
plt.colorbar(im2, ax=axes[0,1], shrink=0.7)

im3 = plot_corr_matrix(axes[1,0], C_diff_ord, "C_IOI − C_nonIOI: what changes",
                        vmin=-0.4, vmax=0.4)
plt.colorbar(im3, ax=axes[1,0], shrink=0.7)

im4 = plot_corr_matrix(axes[1,1], C_delta_ord, "C_ΔKL: all 2500 pairings")
plt.colorbar(im4, ax=axes[1,1], shrink=0.7)

plt.suptitle("Cross-head KL correlation structure\n"
             "Rows/cols: circuit heads by role | non-circuit", fontsize=12, y=1.01)
plt.tight_layout()
plt.savefig("kl_cross_head_correlation_4panel.png", dpi=300, bbox_inches="tight")
plt.show()

# === Zoomed: circuit-only (23x23) ===
fig, axes = plt.subplots(2, 2, figsize=(14, 12))

for ax, mat, title in [
    (axes[0,0], C_ioi_ord[:n_circuit,:n_circuit], "C_IOI (circuit only)"),
    (axes[0,1], C_non_ord[:n_circuit,:n_circuit], "C_nonIOI (circuit only)"),
    (axes[1,0], C_diff_ord[:n_circuit,:n_circuit], "C_IOI − C_nonIOI (circuit only)"),
    (axes[1,1], C_delta_ord[:n_circuit,:n_circuit], "C_ΔKL (circuit only)"),
]:
    vmin, vmax = (-0.4, 0.4) if "−" in title else (-0.6, 0.6)
    im = ax.imshow(mat, cmap='RdBu_r', vmin=vmin, vmax=vmax, aspect='equal')
    ax.set_xticks(range(n_circuit))
    ax.set_xticklabels(ordered_labels[:n_circuit], rotation=90, fontsize=5)
    ax.set_yticks(range(n_circuit))
    ax.set_yticklabels(ordered_labels[:n_circuit], fontsize=5)
    for b in role_boundaries[:-1]:
        ax.axhline(b - 0.5, color='black', lw=1)
        ax.axvline(b - 0.5, color='black', lw=1)
    ax.set_title(title, fontsize=10)
    plt.colorbar(im, ax=ax, shrink=0.7)

plt.suptitle("Circuit heads only (23x23)\nNM | Backup NM | Neg NM | S-Inhib | Induct | Dup Tok | Prev Tok",
             fontsize=11, y=1.01)
plt.tight_layout()
plt.savefig("kl_cross_head_correlation_circuit.png", dpi=300, bbox_inches="tight")
plt.show()

# === Summary stats ===
all_roles = role_order + ["Non-circuit"]

for label, mat in [("C_IOI", C_ioi_ord), ("C_nonIOI", C_non_ord),
                    ("C_IOI - C_nonIOI", C_diff_ord), ("C_ΔKL (2500 pairs)", C_delta_ord)]:
    print(f"\n--- {label} ---")
    print(f"{'Role A':<22} {'Role B':<22} {'mean r':>8}")
    print("-" * 55)
    for i, role_a in enumerate(all_roles):
        for j, role_b in enumerate(all_roles):
            if j < i:
                continue
            idx_a = [k for k, r in enumerate(ordered_roles) if r == role_a]
            idx_b = [k for k, r in enumerate(ordered_roles) if r == role_b]
            pairs = [mat[ia, ib] for ia in idx_a for ib in idx_b if ia != ib]
            if pairs:
                print(f"{role_a:<22} {role_b:<22} {np.mean(pairs):+.3f}")

# === Biggest changes ===
pair_changes = []
for i in range(144):
    for j in range(i+1, 144):
        pair_changes.append((i, j, C_ioi_ord[i,j], C_non_ord[i,j], C_diff_ord[i,j]))

pair_changes.sort(key=lambda x: x[4], reverse=True)

print("\n\n=== Head pairs with largest correlation INCREASE on IOI ===")
print(f"{'Head i':<10} {'Head j':<10} {'C_IOI':>8} {'C_non':>8} {'Δ':>8} {'Role i':<22} {'Role j'}")
print("-" * 90)
for i, j, c_ioi, c_non, diff in pair_changes[:15]:
    print(f"{ordered_labels[i]:<10} {ordered_labels[j]:<10} {c_ioi:+.3f}   {c_non:+.3f}   {diff:+.3f}   "
          f"{ordered_roles[i]:<22} {ordered_roles[j]}")

print("\n=== Head pairs with largest correlation DECREASE on IOI ===")
for i, j, c_ioi, c_non, diff in pair_changes[-15:]:
    print(f"{ordered_labels[i]:<10} {ordered_labels[j]:<10} {c_ioi:+.3f}   {c_non:+.3f}   {diff:+.3f}   "
          f"{ordered_roles[i]:<22} {ordered_roles[j]}")
```

```python
# === Hierarchical clustering on cross-head KL correlation matrices ===
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
from scipy.spatial.distance import squareform

# Reshape: (50, 144)
kl_ioi_mat = kl_ioi.reshape(50, -1)
kl_non_mat = kl_non_ioi.reshape(50, -1)

# Correlation matrices
C_ioi = np.corrcoef(kl_ioi_mat.T)    # (144, 144)
C_non = np.corrcoef(kl_non_mat.T)    # (144, 144)
C_diff = C_ioi - C_non

# Distance = 1 - correlation (standard for correlation clustering)
# Clip to avoid negative distances from numerical noise
D_ioi = np.clip(1 - C_ioi, 0, 2)
D_non = np.clip(1 - C_non, 0, 2)
# For C_diff, values range [-2, 2], so shift: distance = 1 - C_diff
D_diff = np.clip(1 - C_diff, 0, 3)

# Convert to condensed form for scipy
D_ioi_cond = squareform(D_ioi, checks=False)
D_non_cond = squareform(D_non, checks=False)
D_diff_cond = squareform(D_diff, checks=False)

# Linkage (average linkage works well with correlation distance)
Z_ioi = linkage(D_ioi_cond, method='average')
Z_non = linkage(D_non_cond, method='average')
Z_diff = linkage(D_diff_cond, method='average')

# Labels: L{layer}H{head} colored by role
head_labels = [f"L{l}H{h}" for l in range(12) for h in range(12)]
head_roles = [head_role(l, h) for l in range(12) for h in range(12)]

ROLE_COLORS_DEND = {
    "Name Mover": "#d62728",
    "Backup Name Mover": "#ff7f0e",
    "Negative Name Mover": "#9467bd",
    "S-Inhibition": "#2ca02c",
    "Induction": "#17becf",
    "Duplicate Token": "#bcbd22",
    "Previous Token": "#e377c2",
    "Non-circuit": "#999999",
}

def colored_dendrogram(ax, Z, title, labels, roles):
    """Plot dendrogram with leaf labels colored by circuit role."""
    dn = dendrogram(Z, ax=ax, labels=labels, leaf_rotation=90, leaf_font_size=4,
                    color_threshold=0, above_threshold_color='#333333')
    # Color the leaf labels
    xlbls = ax.get_xticklabels()
    for lbl in xlbls:
        head_name = lbl.get_text()
        idx = labels.index(head_name)
        role = roles[idx]
        lbl.set_color(ROLE_COLORS_DEND[role])
        if role != "Non-circuit":
            lbl.set_fontweight('bold')
            lbl.set_fontsize(5)
    ax.set_title(title, fontsize=11)
    ax.set_ylabel("Distance (1 − corr)", fontsize=9)
    return dn

fig, axes = plt.subplots(3, 1, figsize=(20, 18))

colored_dendrogram(axes[0], Z_ioi, "Hierarchical clustering: C_IOI (KL correlations on IOI prompts)",
                   head_labels, head_roles)
colored_dendrogram(axes[1], Z_non, "Hierarchical clustering: C_nonIOI (KL correlations on non-IOI prompts)",
                   head_labels, head_roles)
colored_dendrogram(axes[2], Z_diff, "Hierarchical clustering: C_IOI − C_nonIOI (task-specific coupling)",
                   head_labels, head_roles)

# Add legend
from matplotlib.patches import Patch
legend_elements = [Patch(facecolor=c, label=r) for r, c in ROLE_COLORS_DEND.items() if r != "Non-circuit"]
legend_elements.append(Patch(facecolor="#999999", label="Non-circuit"))
axes[0].legend(handles=legend_elements, loc='upper right', fontsize=7, ncol=2)

plt.tight_layout()
plt.savefig("hierarchical_clustering_dendrograms.png", dpi=300, bbox_inches="tight")
plt.show()

# === Print cluster membership at a few cut thresholds ===
print("\n=== C_IOI clustering: circuit head groupings ===")
for thresh in [0.8, 1.0, 1.2]:
    clusters = fcluster(Z_ioi, t=thresh, criterion='distance')
    print(f"\nThreshold = {thresh}:")
    cluster_ids = sorted(set(clusters))
    for cid in cluster_ids:
        members = [head_labels[i] for i in range(144) if clusters[i] == cid]
        circuit_members = [(head_labels[i], head_roles[i]) for i in range(144)
                          if clusters[i] == cid and head_roles[i] != "Non-circuit"]
        if circuit_members:
            n_total = len(members)
            print(f"  Cluster {cid} ({n_total} heads): "
                  + ", ".join(f"{h}({r})" for h, r in circuit_members)
                  + f" + {n_total - len(circuit_members)} non-circuit")

print("\n\n=== C_diff clustering: task-specific groupings ===")
for thresh in [0.8, 1.0, 1.2]:
    clusters = fcluster(Z_diff, t=thresh, criterion='distance')
    print(f"\nThreshold = {thresh}:")
    cluster_ids = sorted(set(clusters))
    for cid in cluster_ids:
        members = [head_labels[i] for i in range(144) if clusters[i] == cid]
        circuit_members = [(head_labels[i], head_roles[i]) for i in range(144)
                          if clusters[i] == cid and head_roles[i] != "Non-circuit"]
        if circuit_members:
            n_total = len(members)
            print(f"  Cluster {cid} ({n_total} heads): "
                  + ", ".join(f"{h}({r})" for h, r in circuit_members)
                  + f" + {n_total - len(circuit_members)} non-circuit")

# === Where do L8H1 and L8H11 cluster relative to circuit heads? ===
print("\n\n=== L8H1 and L8H11: cluster neighbors ===")
for target in ["L8H1", "L8H11"]:
    target_idx = head_labels.index(target)
    for label, Z in [("C_IOI", Z_ioi), ("C_diff", Z_diff)]:
        clusters = fcluster(Z, t=1.0, criterion='distance')
        target_cluster = clusters[target_idx]
        neighbors = [(head_labels[i], head_roles[i]) for i in range(144)
                     if clusters[i] == target_cluster and i != target_idx]
        circuit_neighbors = [f"{h}({r})" for h, r in neighbors if r != "Non-circuit"]
        n_non = sum(1 for _, r in neighbors if r == "Non-circuit")
        print(f"{target} in {label} (t=1.0): cluster {target_cluster} — "
              f"circuit neighbors: {circuit_neighbors or 'none'}, "
              f"+ {n_non} non-circuit")
```

```python
# === Reorder correlation matrices by C_diff dendrogram leaf order ===
# This should reveal block structure aligned with task-specific functional coupling

from scipy.cluster.hierarchy import leaves_list

# Get leaf order from C_diff dendrogram
diff_leaf_order = leaves_list(Z_diff)

# Reorder all matrices by this order
C_ioi_dord = C_ioi[np.ix_(diff_leaf_order, diff_leaf_order)]
C_non_dord = C_non[np.ix_(diff_leaf_order, diff_leaf_order)]
C_diff_dord = C_diff[np.ix_(diff_leaf_order, diff_leaf_order)]

# Labels and roles in dendrogram order
labels_dord = [head_labels[i] for i in diff_leaf_order]
roles_dord = [head_roles[i] for i in diff_leaf_order]

# Color bar for roles
role_color_array = [ROLE_COLORS_DEND[r] for r in roles_dord]

fig, axes = plt.subplots(1, 3, figsize=(22, 7))

for ax, mat, title, vlim in [
    (axes[0], C_ioi_dord, "C_IOI", 0.6),
    (axes[1], C_non_dord, "C_nonIOI", 0.6),
    (axes[2], C_diff_dord, "C_IOI − C_nonIOI", 0.5),
]:
    im = ax.imshow(mat, cmap='RdBu_r', vmin=-vlim, vmax=vlim, aspect='equal')
    ax.set_title(title, fontsize=12)
    plt.colorbar(im, ax=ax, shrink=0.7)
    
    # Add colored role ticks on both axes
    for idx_pos in range(144):
        color = role_color_array[idx_pos]
        if color != "#999999":  # only mark circuit heads
            ax.plot(-2, idx_pos, 's', color=color, markersize=3, clip_on=False)
            ax.plot(idx_pos, 145, 's', color=color, markersize=3, clip_on=False)
    
    # Label circuit heads on y-axis
    circuit_ticks = [(i, labels_dord[i]) for i in range(144) if roles_dord[i] != "Non-circuit"]
    ax.set_yticks([i for i, _ in circuit_ticks])
    ax.set_yticklabels([l for _, l in circuit_ticks], fontsize=4)
    ax.set_xticks([])

# Legend
from matplotlib.patches import Patch
legend_elements = [Patch(facecolor=c, label=r) for r, c in ROLE_COLORS_DEND.items()]
axes[2].legend(handles=legend_elements, loc='upper right', fontsize=6, 
               bbox_to_anchor=(1.35, 1.0))

plt.suptitle("Cross-head KL correlations — ordered by C_diff dendrogram", fontsize=13, y=1.02)
plt.tight_layout()
plt.savefig("correlation_matrices_dendrogram_ordered.png", dpi=300, bbox_inches="tight")
plt.show()

# Print the dendrogram-ordered circuit heads (to see grouping)
print("\nCircuit heads in C_diff dendrogram order:")
for i, idx in enumerate(diff_leaf_order):
    role = head_roles[idx]
    if role != "Non-circuit":
        print(f"  position {i:3d}: {head_labels[idx]:<8} ({role})")
```

## Top-50 C_diff Pairs

Pairs of heads ranked by |C_diff|. Circuit–circuit (CC) pairs are expected;
circuit–non-circuit (CN) pairs are the interesting discovery signal.
Non-circuit–non-circuit (NN) pairs that rank highly are noise/confounds.

```python
# === Top-50 |C_diff| pairs ===
hl_flat = [(l, h) for l in range(12) for h in range(12)]
cmask_flat = np.array([(l, h) in ALL_CIRCUIT_HEADS for l, h in hl_flat])

def _role(l, h):
    for r, hs in IOI_HEADS.items():
        if (l, h) in hs: return r
    return 'Non-circuit'

pairs_ranked = sorted(
    [(abs(C_diff_lyr[i, j]), C_diff_lyr[i, j], i, j)
     for i in range(144) for j in range(i+1, 144)],
    reverse=True
)
top50 = pairs_ranked[:50]

heads_seen = set()
for _, _, i, j in top50:
    heads_seen.add(hl_flat[i]); heads_seen.add(hl_flat[j])
circuit_seen    = heads_seen & ALL_CIRCUIT_HEADS
circuit_missing = ALL_CIRCUIT_HEADS - heads_seen
print(f'Top-50 |C_diff| pairs cover {len(circuit_seen)}/{len(ALL_CIRCUIT_HEADS)} circuit heads')
print(f'Missing: {sorted(circuit_missing)}')

# Bar chart
COLORS_CAT = {'CC': '#d62728', 'CN': '#ff7f0e', 'NN': '#bbbbbb'}
fig2, ax = plt.subplots(figsize=(13, 4))
for k, (av, v, i, j) in enumerate(top50):
    ic = cmask_flat[i]; jc = cmask_flat[j]
    ct = 'CC' if ic and jc else ('CN' if ic or jc else 'NN')
    ax.bar(k, v, color=COLORS_CAT[ct], width=0.8, linewidth=0)

# Annotate CC bars
for k, (av, v, i, j) in enumerate(top50):
    ic = cmask_flat[i]; jc = cmask_flat[j]
    if ic and jc:
        li, hi = hl_flat[i]; lj, hj = hl_flat[j]
        ax.text(k, (v if v >= 0 else 0) + 0.02,
                f'L{li}H{hi}\nL{lj}H{hj}',
                ha='center', va='bottom', fontsize=5, color='#990000', rotation=90)

ax.axhline(0, color='black', lw=0.8)
ax.set_xlabel('Rank (by |C_diff,ij|)', fontsize=11)
ax.set_ylabel('C_diff,ij (signed)', fontsize=11)
ax.set_title(
    f'Top 50 pairs by |C_diff|  —  covers {len(circuit_seen)}/{len(ALL_CIRCUIT_HEADS)} known circuit heads\n'
    f'Missing: {", ".join(f"L{l}H{h}" for l, h in sorted(circuit_missing)) or "none"}',
    fontsize=10)
ax.set_xlim(-1, 50)

from matplotlib.patches import Patch
ax.legend(handles=[Patch(facecolor=COLORS_CAT[c], label=c) for c in ['CC', 'CN', 'NN']],
          fontsize=9, loc='upper right')
ax.grid(True, alpha=0.2, axis='y')
plt.tight_layout()
plt.show()

# Table
print(f'\n{"Rk":<4} {"Head i":<8} {"Role i":<22} {"Head j":<8} {"Role j":<22} {"C_diff":>7}  cat')
print('-'*80)
for k, (av, v, i, j) in enumerate(top50, 1):
    li, hi = hl_flat[i]; lj, hj = hl_flat[j]
    ic = cmask_flat[i]; jc = cmask_flat[j]
    ct = 'CC' if ic and jc else ('CN' if ic or jc else 'NN')
    print(f'{k:<4} L{li}H{hi:<5} {_role(li,hi):<22} L{lj}H{hj:<5} {_role(lj,hj):<22} {v:+.3f}  {ct}')
```
